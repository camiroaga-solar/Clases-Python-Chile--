# 4) Entrada y salida en Python

## Input

Ya conocemos bien la funciÃ³n **"print"**. Lo podemos usar para mostrar resultados en la pantalla. Visto de otra forma, es una manera de que datos "salgan" de Python, para que puedan ser vistos por nosotros.

**"Input"** es otra funciÃ³n que puede ser vista como el inverso de "print", pues la usamos para que "entren" datos a Python.

VeÃ¡mosla en acciÃ³n:

```python
nombre = input("Ingresa tu nombre: ")
print("Hola,", nombre, "!")
```

Python recibiÃ³ el nombre, lo guardÃ³ en la variable nombre, y luego lo imprimiÃ³ junto con un "hola".

El texto que va dentro de los parÃ©ntesis del "input" es el texto con el cual se solicita al usuario el dato pedido.

Otro ejemplo:

```python
edad = input("Ingresa tu edad: ")
print("Tengo", edad, "aÃ±os")
```

Usemos dos "input" para recibir nombres y apellidos:

```python
nombre = input("Ingresa tu nombre: ")
apellido = input("Ingresa tu apellido: ")
print("Hola", nombre, apellido, "!")
```

Veamos si podemos usar esto para calcular el doble de tu edad:

```python
edad = input("Ingresa tu edad: ")
doble = edad * 2
print("El doble de mi edad es", doble, "aÃ±os")
```

Â¿QuÃ© pasÃ³? ðŸ˜¨

Input() por default solo recibe texto o strings.

Por lo tanto, Python multiplicÃ³ por dos el texto, no el nÃºmero.

Si queremos hacer operaciones sobre el nÃºmero, tenemos que aÃ±adir un "int()", que es una funciÃ³n que convierte objetos en nÃºmeros enteros:

```python
edad = input("Ingresa tu edad: ")

edad_int = int(edad) # AcÃ¡ convertimos a int!

doble = edad_int * 2
print("El doble de mi edad es", doble, "aÃ±os")
```
