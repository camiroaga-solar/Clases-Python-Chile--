# 4) Entrada y salida en Python

## Input

Ya conocemos bien la funci√≥n **"print"**. Lo podemos usar para mostrar resultados en la pantalla. Visto de otra forma, es una manera de que datos "salgan" de Python, para que puedan ser vistos por nosotros.

**"Input"** es otra funci√≥n que puede ser vista como el inverso de "print", pues la usamos para que "entren" datos a Python.

Ve√°mosla en acci√≥n:

```python
nombre = input("Ingresa tu nombre: ")
print("Hola,", nombre, "!")
```

Python recibi√≥ el nombre, lo guard√≥ en la variable nombre, y luego lo imprimi√≥ junto con un "hola".

El texto que va dentro de los par√©ntesis del "input" es el texto con el cual se solicita al usuario el dato pedido.

Otro ejemplo:

```python
edad = input("Ingresa tu edad: ")
print("Tengo", edad, "a√±os")
```

Usemos dos "input" para recibir nombres y apellidos:

```python
nombre = input("Ingresa tu nombre: ")
apellido = input("Ingresa tu apellido: ")
print("Hola", nombre, apellido, "!")
```

Veamos si podemos usar esto para calcular el doble de tu edad:

```python
edad = input("Ingresa tu edad: ")
doble = edad * 2
print("El doble de mi edad es", doble, "a√±os")
```

¬øQu√© pas√≥? üò®

Input() por default solo recibe texto o strings.

Por lo tanto, Python multiplic√≥ por dos el texto, no el n√∫mero.

Si queremos hacer operaciones sobre el n√∫mero, tenemos que a√±adir un "int()", que es una funci√≥n que convierte objetos en n√∫meros enteros:

```python
edad = input("Ingresa tu edad: ")

edad_int = int(edad) # Ac√° convertimos a int!

doble = edad_int * 2
print("El doble de mi edad es", doble, "a√±os")
```
