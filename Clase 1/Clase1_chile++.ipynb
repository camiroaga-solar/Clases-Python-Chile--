No te preocupes en entender el c√≥digo, pero veamos qu√© pasa cuando apretamos el bot√≥n de "Ejecutar".

```python
import random

print("¬°Bienvenido al juego de adivinanzas!")
print("Estoy pensando en un n√∫mero entre 1 y 100.")

numero_secreto = random.randint(1, 100)
intentos = 0

while True:
    intentos += 1
    suposicion = int(input("Ingresa tu suposici√≥n: "))

    if suposicion < numero_secreto:
        print("Demasiado bajo. Intenta de nuevo.")
    elif suposicion > numero_secreto:
        print("Demasiado alto. Intenta de nuevo.")
    else:
        print("¬°Felicitaciones! ¬°Adivinaste el n√∫mero en", intentos, "intentos!")
        break

print("Gracias por jugar.")
```

Veamos este otro. Intenta hacer click en los botones


```python
from ipywidgets import Button, Layout, VBox, Output
from IPython.display import display

def on_button_clicked(b):
    with output:
        output.clear_output()
        if b.description == '¬°Hola!':
            print("¬°Hola, minion!")
        elif b.description == 'Adi√≥s':
            print("Chao minion.")


button_hello = Button(description='¬°Hola!', button_style='success', layout=Layout(height='auto', width='auto'))
button_bye = Button(description='Adi√≥s', button_style='danger', layout=Layout(height='auto', width='auto'))
output = Output()
button_hello.on_click(on_button_clicked)
button_bye.on_click(on_button_clicked)

display(VBox([button_hello, button_bye, output]))
```

Veamos este otro. De nuevo, no te preocupes por el c√≥digo, pero experimenta con los sliders que dicen "R,G,y B". ¬øQu√© significan?

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from ipywidgets import interactive

# Funci√≥n para mostrar un rect√°ngulo con un color basado en valores RGB
def dibujar_rectangulo_colorido(R=255, G=255, B=255):
    # Crear una nueva figura y ejes
    fig, ax = plt.subplots(figsize=(6, 3))

    # A√±adir un rect√°ngulo a los ejes
    rect = Rectangle((0.1, 0.1), 0.8, 0.8, color=(R/255, G/255, B/255))
    ax.add_patch(rect)

    # Establecer l√≠mites y apagar los ejes para una mejor visualizaci√≥n
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')

    plt.show()

# Crear controles deslizantes interactivos para cada componente RGB
plot_interactivo = interactive(dibujar_rectangulo_colorido, R=(0, 255), G=(0, 255), B=(0, 255))
salida = plot_interactivo.children[-1]
salida.layout.height = '250px' # Ajustar la altura de la trama de salida

# Mostrar el widget interactivo
plot_interactivo

```

¬°Interesante! **R, G y B** significan "Red", "Green" y "Blue", y nos permiten generar todos los colores a trav√©s de la combinaci√≥n de estos tres.
